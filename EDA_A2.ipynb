```python
# Question 2.5:

price_log_beds = df.plot.scatter(x='price_log',y='Beds')
price_log_beds.set(title='Log of Price and Beds',xlabel='Log Price',ylabel='Beds')
# The scatterplot shows that the number of beds tends toward the 0-6 range at a median price range. 

var = 'Price'
group = 'Beds'
df.loc[:,[group,var]].groupby(group).describe()
# The description shows that the mean price for renting loosely increases as the number of beds increases. The standard deviation is quite large,
# indicating a large spread in the data, which likely accounts for the lack of a clear increase in price. 
```

```python
var = 'Price'
group2 = 'Property Type'
df.loc[:,[group2,var]].groupby(group2).describe()
# When comparing price with room type, renting a whole home/apartment is by far the highest price as well as standard deviation. When comparing 
# property type with price, afte "other", the highest renting option is a villa. Standard deviation shows the same trend. 

# Median is a more reliable estimate of central tendency compared to the mean because it is not as impacted by outliers. In this dataset especially, 
# the standard deviation demonstrates that the spread of data is quite large, causing the means to be skewed. The median, however, is less impacted
# by this skew, and is therefore a better measure of central tendency. 
```

```python
# Question 3.1: Load ./data/drilling_rigs.csv and examine the data. How many observations? How many variables? Are numeric variables correctly read in 
# by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.

df2 = pd.read_csv("data/drilling_rigs.csv")

print(df2.shape)
print(df2.columns)
print(df2.head(3))
print(df2.tail(2))
print(df2.dtypes)

# There are 623 observations and 10 variables in this dataset. Not all the numeric variables are correctly red in by Pandas. There are strings 
# such as "Not Available" that will need to be replaced using the .replace function using np.nan so that they appear as NaN, allowing for the rest of 
# the data to be coerced into numerical data. There doesn't appear to be any issues with numbers greater than 999 containing ",". 
```

```python
# Question 3.2: 

df2["time"]=pd.to_datetime(df2["Month"],format = 'mixed')
# print(df2.head(2))
```

```python
# Question 3.3: 

#print(df2.dtypes)
#print(df2["Active Well Service Rig Count (Number of Rigs)"].unique())
df2["Active Well Service Rig Count (Number of Rigs)"]=df2["Active Well Service Rig Count (Number of Rigs)"].replace("Not Available",np.nan)
df2["Active Well Service Rig Count (Number of Rigs)"] = df2["Active Well Service Rig Count (Number of Rigs)"].astype(float)
rigs = df2["Active Well Service Rig Count (Number of Rigs)"]
rigs_plot = rigs.plot(kind="line",title = "Number of Rigs")
rigs_plot.set(title='Number of Rigs',xlabel='Count',ylabel='Total Number of Rigs over Time')

# The number of rigs appears to increase and then decrease. However, the trend is far from linear and is marked by dramatic peaks and dips. 
```

```python
# Question 3.4: 

rig_overtime = sns.lineplot(data=df2,x='time',y='Active Well Service Rig Count (Number of Rigs)')
rig_overtime.set(title='Number of Rigs Over Time',xlabel='Time',ylabel='Total Number of Rigs')

# The total number of rigs over time decreases over time. There are, however, many outliers as shown by the various peaks. The graph looks the same as
# question 3.3 because the data's position in the dataframe was already chronological. Thus, adding the time variable did not actually change the shape
# of the graph, it simply changed the values marking the x axis. 
```
